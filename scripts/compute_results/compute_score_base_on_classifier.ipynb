```python
import json
import numpy as np
from scipy.special import softmax

from scripts import config
```

```python
with open(config.CLASSIFIER_SCORES_PATH, "r", encoding="utf-8") as f:
    results_cl = json.load(f)
```

```python
# We need to get the probability of cartoon based on the results
results_prob = {}
for run_id in results_cl:
    for epoch in results_cl[run_id]:
        all_probs = []
        for result_classifier in results_cl[run_id][epoch]:
            all_probs.append(softmax(result_classifier)[1])
        results_prob[f"{run_id}_{epoch}"] = np.mean(all_probs)
```

```python
# Now we can get the results
results = sorted([(k, v) for (k, v) in results_prob.items()], key=lambda x: x[1], reverse=True)
results
```

```python
to_remove = ["2022_03_31-18_37_59_gpK","2022_04_04-18_27_57_veL","2022_04_04-03_19_26_QAj","2022_03_31-18_38_39_AqQ","2022_03_31-18_56_57_xfY","2022_04_01-04_25_45_DEW","2022_04_01-18_56_08_GUY","2022_04_03-20_04_09_Bkv","2022_04_03-20_03_58_Sko"]
results_filtered = []

for res in results:
    should_remove = False
    for r in to_remove:
        if res[0].startswith(r):
            should_remove = True
    if not should_remove:
        results_filtered.append(res)

results_filtered
```
